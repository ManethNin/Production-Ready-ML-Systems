```python
df = pd.read_csv('processed/CEHHbInToW_outlier_handled.csv')
df.head()
```

```python
def creditScoreIntoBins(score):
    if 350<=score<=580:
        return 'Poor'
    elif 580<score<=670:
        return 'Fair'
    elif 670<score<=740:
        return 'Good'
    elif 740<score<=800:
        return 'Very Good'
    elif 800<score<=850:
        return 'Excellent'
    else:
        print(f'Invalid credit score: {score}')
        assert True, f'Invalid credit score: {score}'

df['credit Score Bins'] = df['CreditScore'].apply(creditScoreIntoBins)
```
